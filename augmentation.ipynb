```python
import os
import matplotlib.pyplot as plt
import albumentations as A
import cv2
```

```python
folder = 'CK+'
emotion = 'anger'
folder_path = os.path.join(folder, emotion)
```

```python
images = []
names = []

for img_name in os.listdir(folder_path):
    img_path = os.path.join(folder_path, img_name)
    # convert("L") converts the image to greyscale
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    names.append(img_name)
    images.append(img)
```

```python
print('Liczba zdjeć: ', len(images))
```

### Transformacja z Albumentations

```python
transform = A.Compose([
    A.Crop(x_min=200, x_max=500, y_min=15, y_max=415)
])
```

```python
augmented_image = transform(image=images[0])['image']
```

```python
fig, ax = plt.subplots(1, 2, figsize=[12, 4])
ax[0].imshow(images[0], cmap='gray')
ax[0].set_title('Original image')
ax[1].imshow(augmented_image, cmap='gray')
ax[1].set_title('Transformed')
```

```python
augmented = []
for i in range(10):
    augmented.append(transform(image=images[i])['image'])
```

```python
fig, ax = plt.subplots(10, 2, figsize=[15, 40])
for i in range(10):
    ax[i, 0].imshow(images[i], cmap='gray')
    ax[i, 1].imshow(augmented[i], cmap='gray')
```
